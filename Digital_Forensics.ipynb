```python
img = cv2.imread('hand.jpg')

h,w = img.shape[:2]

start_row, start_colm = int(h*0.45), int(w*0.35)
end_row, end_colm = int(h*0.76), int(w*0.76)

cropped_img =img[start_row:end_row, start_colm:end_colm]

#cv2.imshow('ROI',cropped_img)

plt.imshow(cropped_img)
plt.show()
```

```python
cv2.waitKey()
cv2.destroyAllWindows() #close all windows bhi hota hai
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
#threshold filter and binary filter
img=cv2.imread("hand.jpg", 0)

ret, th=cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

ret, bn = cv2.threshold(img, 127,255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU) # to invert the OTSU

#cv2.imshow('original', img)
plt.imshow(img)
plt.show()

#cv2.imshow("threshold", th)
plt.imshow(th)
plt.show()


#cv2.imshow("Invert binary", bn)
plt.imshow(bn)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("hand.jpg")
#laplacian filter edge detection filter
lap_img=cv2.Laplacian(img, cv2.CV_64F, ksize=3)#3*3 filte k size
#cv2.imshow('original', img)
plt.imshow(img)
plt.show()

#cv2.imshow("Laplacian", lap_img)
plt.imshow(lap_img)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("hand.jpg")

kernel=np.ones((5,5), np.float32)/80
lpf_img=cv2.filter2D(img, -1, kernel)
#cv2.imshow("original", img)
plt.imshow(img)
plt.show()

#cv2.imshow("LPF img", lpf_img)
plt.imshow(lpf_img)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("hand.jpg")

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define a high-pass filter kernel (3x3 Laplacian kernel)
kernel_hp = np.array([[0, -1, 0],
                      [-1, 4, -1],
                      [0, -1, 0]], np.float32)

# Apply the high-pass filter
hpf_img = cv2.filter2D(gray_img, -1, kernel_hp)

# Display original image
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Display high-pass filtered image
plt.subplot(1, 2, 2)
plt.imshow(hpf_img, cmap='gray')
plt.title('High-Pass Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
